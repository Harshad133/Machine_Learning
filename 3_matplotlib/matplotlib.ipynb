```python
data = pd.read_csv('sharma-kohli.csv')
data
```

```python
# plot 2 columns using simple plot 
plt.plot(data['index'], data['V Kohli'])
```

```python
# plot multiple graphs

plt.plot(data['index'], data['V Kohli'])
plt.plot(data['index'], data['RG Sharma'])
```

```python
# add title and lebels

plt.plot(data['index'], data['V Kohli'])
plt.plot(data['index'], data['RG Sharma'])

plt.title('Sharma vs Kohli')
plt.xlabel('Year')
plt.ylabel('Runs')
```

```python
# change colors

plt.plot(data['index'], data['V Kohli'], color='black')
plt.plot(data['index'], data['RG Sharma'], color='green')

plt.title('Sharma vs Kohli')
plt.xlabel('Year')
plt.ylabel('Runs')
```

```python
# line style and line width

plt.plot(data['index'], data['V Kohli'], color='black', linestyle='dashed', linewidth=3)
plt.plot(data['index'], data['RG Sharma'], color='green', linestyle='dotted')

plt.title('Sharma vs Kohli')
plt.xlabel('Year')
plt.ylabel('Runs')
```

```python
# datapoint is called marker and marker size

plt.plot(data['index'], data['V Kohli'], color='black', linestyle='dashed', linewidth=3, marker='o')
plt.plot(data['index'], data['RG Sharma'], color='green', linestyle='dotted', marker='*', markersize=10)

plt.title('Sharma vs Kohli')
plt.xlabel('Year')
plt.ylabel('Runs')
```

```python
# legend 
# show label of line in legende

plt.plot(data['index'], data['V Kohli'], color='black', linestyle='dashed', linewidth=3, marker='o', label='Virat')
plt.plot(data['index'], data['RG Sharma'], color='green', linestyle='dotted', marker='*', markersize=10 , label='Rohit')

plt.title('Sharma vs Kohli')
plt.xlabel('Year')
plt.ylabel('Runs')

# labale is not show untill call plt.legende()

plt.legend()
# can adjust postion of legende using loc parameter
```

```python
# using plt.xlim and plt.ylim we can set limit of x and y axis

plt.plot(data['index'], data['V Kohli'], color='black', linestyle='dashed', linewidth=3, marker='o', label='Virat')
plt.plot(data['index'], data['RG Sharma'], color='green', linestyle='dotted', marker='*', markersize=10 , label='Rohit')

plt.title('Sharma vs Kohli')
plt.xlabel('Year')
plt.ylabel('Runs')

plt.legend()

plt.xlim(2010,2015)
plt.ylim(0,1000)

plt.grid()
plt.show() # show graph
```

# Bar chart

-> bivariante analsys
-> numerical -> categorical
-> aggreagte analysys

```python
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'], df['2015'])
```

```python
# plot multiple bar graph in one plot
# when this do will overlap bars

plt.bar(df['batsman'], df['2015'], color='red', label='2015')
plt.bar(df['batsman'], df['2016'], color='green', label='2016')
plt.bar(df['batsman'], df['2017'], color='blue', label='2017')

# like this

```

```python
# width parameter is decide the width of bar

# plot multiple bar 

plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2)
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2)
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2)

# change label of axis/ change the x cordinates
plt.xticks(np.arange(df.shape[0]), df['batsman'])
```

```python
# label is long that time use this 

plt.xticks(rotation='vertical')
```

```python
# Stacked bar chart

plt.bar(df['batsman'], df['2015'], color='red', label='2015')
plt.bar(df['batsman'], df['2016'], color='green', label='2016', bottom=df['2015'])
plt.bar(df['batsman'], df['2017'], color='blue', label='2017', bottom=df['2015']+df['2016'])

plt.legend()
```

Pia chart

-> univariante and biaveriante analysis
-> categorical and numerical 
-> to find contribution on standard scale

```python
df = pd.read_csv('gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%1.1f%%')
```

```python
# color
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%1.1f%%', colors=['red', 'green', 'blue', 'cyan', 'magenta', 'yellow'])
```

```python
# explode some category
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%1.1f%%', explode=[1, 0, 0.1, 0.1, 0, 0])
```
