```python
# fetch all final winner

i = ipl_matches['MatchNumber'] == 'Final'
new = ipl_matches[i]
new[['Season','WinningTeam']]

ipl_matches[ipl_matches['MatchNumber'] == 'Final'][['Season','WinningTeam']]
```

```python
# Fetch all Super over match

ipl_matches[ipl_matches['SuperOver'] == 'Y'][['Season','Team1','Team2']]
```

```python
# how many matches CSK won in Kolkata

ipl_matches[(ipl_matches['City'] == 'Kolkata') & (ipl_matches['WinningTeam'] == 'Chennai Super Kings')].shape[0]
```

```python
# toss winner is match winner
ipl_matches[ipl_matches['TossWinner'] == ipl_matches['WinningTeam']].shape[0]/ipl_matches.shape[0]*100
```
