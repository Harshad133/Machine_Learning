## SMOTE
SMOTE (Synthetic Minority Oversampling Technique) is a machine learning technique used to address class imbalance in datasets. It works by generating synthetic samples of the minority class, effectively increasing the number of minority class instances in the dataset. 

#### How it Works:
SMOTE selects minority class samples that are close to each other in the feature space. 
It then draws a line between these selected samples and generates a new sample at a point along that line. 
This new sample is synthetic, meaning it doesn't exist in the original dataset. 

```python
from sklearn.datasets import make_classification

X,y = make_classification(n_samples=1000, n_features=2, n_redundant=0, n_clusters_per_class=1, random_state=12, weights=[0.90])
```

```python
import pandas as pd
df = pd.DataFrame(X, columns=['feature1', 'feature2'])
df['target'] = y
df.head()
```

```python
df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['feature1'], df['feature2'], c=df['target'], cmap='coolwarm', alpha=0.5)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Synthetic Classification Dataset')
plt.show()
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
```

```python
X,y = smote.fit_resample(df[['feature1', 'feature2']], df['target'])
df_balanced = pd.DataFrame(X, columns=['feature1', 'feature2'])
df_balanced['target'] = y
```

```python
df_balanced['target'].value_counts()
```

```python
plt.scatter(df_balanced['feature1'], df_balanced['feature2'], c=df_balanced['target'], cmap='coolwarm', alpha=0.5)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Balanced Synthetic Classification Dataset')
plt.show()
```
