```python
# Series from List

data = ["India", "Nepal", "USA", "UK"]

pd.Series(data)
# dtype object means string
```

```python
# integer
data = [1,2,3,4,5]

pd.Series(data)
```

```python
# with custom index

data = ["India", "Nepal", "USA", "UK"]
index = [22,33,44,55]

pd.Series(data, index=index, name="data")
```
